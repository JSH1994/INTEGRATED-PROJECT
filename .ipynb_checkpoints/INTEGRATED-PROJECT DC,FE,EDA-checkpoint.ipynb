<h2>Loading Data</h2>

```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
from sklearn.preprocessing import OneHotEncoder
import gc

# Read the manually cleaned Excell data
file_path = "insurance_claims_cleaned.xlsx"
df = pd.read_excel(file_path)
df
```

<h2>feature engineering</h2>

```python
# Choose the column for the histogram
column_name = 'age'

# Plot the histogram
plt.hist(df[column_name], bins=3, edgecolor='black')

# Add labels and title
plt.xlabel(column_name)
plt.ylabel('Frequency')
plt.title(f'Histogram of {column_name}')

# Display the histogram
plt.show()
```

```python
bin_edges = [0, 30, 55, 100]  # Define the bin edges
bin_labels = ['Young Adult', 'Middle Aged', 'Elderly']  # Corresponding labels for each bin

# Create a new column based on the bin labels
df['ages_category'] = pd.cut(df['age'], bins=bin_edges, labels=bin_labels)
```

```python
df["Contract Years"] = df["months_as_customer"]/12
```

```python
df['total_premiums_paid'] = (df['policy_annual_premium']/12) * df['months_as_customer']
```

```python
df['net_value_of_customer'] = df['total_premiums_paid'] - df['total_claim_amount']
```

```python
# Extract year from the 'policy_bind_date' column
df['policy_bind_year'] = pd.to_datetime(df['policy_bind_date']).dt.year

# Remove the original 'policy_bind_date' column
df.drop(columns=['policy_bind_date'], inplace=True)
```

```python
# Extract week number, month, and day of the week
df['incident_week'] = df['incident_date'].dt.isocalendar().week
df['incident_month'] = df['incident_date'].dt.month
df['incident_day_of_week'] = df['incident_date'].dt.dayofweek + 1
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Calculate relative risk
risk = ((df['total_claim_amount'] - (df['months_as_customer'] / 12 * df['policy_annual_premium'])) / df['total_claim_amount'])

# Scale the relative risk values using StandardScaler (Z-score scaling)
scaler = StandardScaler()
df['scaled_risk'] = scaler.fit_transform(risk.values.reshape(-1, 1))
```

```python
# Calculate the age at policy inception 
df['age_at_policy_inception'] = df['age'] - (df['months_as_customer'] / 12)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_category_distribution(data, column_name, plot_title, shift_right=0.5):
    """
    Plot the distribution of categories from a specified column in a dataset.
    
    Parameters:
    - data: pandas DataFrame, the dataset containing the category column
    - column_name: str, the name of the column containing categories as text
    - plot_title: str, the title of the plot
    - shift_right: float, the value by which to shift the percentage labels to the right
    """
    
    # Explode the category column into multiple rows if it contains comma-separated values
    category_series = data[column_name]

    # Calculate the count for each category
    category_counts = category_series.value_counts()

    # Convert the counts to percentages
    category_pct = (category_counts / category_counts.sum()) * 100

    # Sort the counts from least to most popular
    category_pct = category_pct.sort_values()

    # Create the plot
    plt.figure(figsize=(12, 6))
    ax = sns.barplot(x=category_pct, y=category_pct.index, palette='viridis')

    # Customize the plot to match the theme
    plt.xlabel('Percentage of clients', fontsize=15, color='blue')
    plt.ylabel(None, color='blue')
    plt.title(plot_title, fontsize=15, color='blue')
    plt.xticks(rotation=45)

    # Add percentage labels on each bar
    for index, value in enumerate(category_pct):
        ax.text(value + shift_right, index, f'{value:.2f}%', color='black', ha="center", va="center")
    
    plt.show()
```

```python
plot_category_distribution(df, 'insured_hobbies', "Hobies by Popularity", 0.17625)
```
